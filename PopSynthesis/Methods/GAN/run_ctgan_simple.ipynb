```python
from ctgan import CTGAN
import pandas as pd
```

```python
seed_data = pd.read_csv("data/pp_sample_ipu.csv")
seed_data = seed_data.drop(columns=["serialno", "sample_geog"])
```

```python
# Names of the columns that are discrete
discrete_columns = seed_data.columns.to_list()

ctgan = CTGAN(epochs=1000)
ctgan.fit(seed_data, discrete_columns)
syn_pop = ctgan.sample(6305666)
```

```python
syn_pop.to_csv("output/pp_synthetic_ctgan.csv", index=False)
```
